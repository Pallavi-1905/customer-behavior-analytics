```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('customer_behavior_data.csv')
df.head()
```

```python
# Basic Info
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Visualization - Age Distribution
plt.hist(df['age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# ROI Analysis
high_roi = df[df['roi'] > 0]
low_roi = df[df['roi'] <= 0]

print('High ROI Customers:', len(high_roi))
print('Low ROI Customers:', len(low_roi))
```
